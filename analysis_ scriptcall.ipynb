```python
import pandas as pd
from analysis_utils import (
    run_regression_analysis,   # scenario/regression
    run_correlation_analysis,  # correlation
    run_hypothesis_test        # hypothesis testing
)

# ✅ Step 1: Choose file manually
file_path = input("Enter path to your dataset file (CSV/Excel): ")
if file_path.endswith(".csv"):
    df = pd.read_csv(file_path)
else:
    df = pd.read_excel(file_path)

print("Available columns:", df.columns.tolist())

# ✅ Step 2: Ask which analysis to run
analysis_type = input("Which analysis do you want to run? (regression/correlation/hypothesis): ").strip().lower()

# ✅ Step 3: Ask for predictors/targets depending on analysis
if analysis_type == "regression" or analysis_type == "scenario":
    predictors = input("Enter predictor columns (comma-separated): ").split(",")
    predictors = [col.strip() for col in predictors]
    target = input("Enter target column: ").strip()
    results = run_regression_analysis(df, predictors, target)

elif analysis_type == "correlation":
    predictors = input("Enter predictor columns (comma-separated): ").split(",")
    predictors = [col.strip() for col in predictors]
    target = input("Enter target column: ").strip()
    results = run_correlation_analysis(df, predictors, target)

elif analysis_type == "hypothesis":
    group_col = input("Enter grouping column (e.g., campaign, category): ").strip()
    target = input("Enter target column: ").strip()
    results = run_hypothesis_test(df, group_col, target)

else:
    raise ValueError("Invalid choice. Please select regression, correlation, or hypothesis.")

# ✅ Step 4: Print results in DataFrame
print("\nAnalysis Output:")
print(pd.DataFrame(results))
```
